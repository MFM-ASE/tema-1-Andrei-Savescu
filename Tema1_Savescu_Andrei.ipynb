# Randamente Coca-Cola VS Pepsi

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
COLA = yf.Ticker('KO')
df = COLA.history(period= '1d',start='2019-1-1',end='2020-3-1')
```

```python
df.info
```

```python
df
```

```python
PEP = yf.Ticker('PEP')
df2 = PEP.history(period= '1d',start='2019-1-1',end='2020-3-1')
```

```python
df2
```

```python
close = df["Close"]

simple_return = close.pct_change()

log_return = np.log(1+simple_return)
```

```python
pylab.rcParams['figure.figsize'] = (15, 10)
simple_return.plot()
log_return.plot() 

```

```python
close2 = df2["Close"]

simple_return2 = close2.pct_change()

log_return2 = np.log(1+simple_return2)
```

```python
simple_return2.plot()
log_return2.plot()
```

```python
plt.plot(simple_return,color='red')
plt.plot(simple_return2,color='blue')
plt.title("Randamente",fontsize=20)
plt.legend(['Coca-Cola', 
            'Pepsi'])
plt.show()
```

```python
plt.hist(simple_return, bins=25, alpha=0.45, color='red')
plt.hist(simple_return2, bins=25, alpha=0.45, color='blue')
plt.title("Distributia randamentelor",fontsize=20)
plt.legend(['Coca-Cola', 
            'Pepsi'])
plt.show()
```
